```python
iris = load_iris() ## Initiate 
```

```python
data_iris = iris.data  ## picking up data only or calling
```

```python
data_iris ## Variables
```

```python
iris_features = pd.DataFrame(data_iris, columns=['sepal length', 'sepal width', 'pedal length', 'pedal width'])

iris_features.head()
```

```python
iris_target = iris.target
```

```python
target_df = pd.DataFrame(iris_target, columns=['target'])
target_df.head()
```

```python
## concat 

data_iris = pd.concat([iris_features, target_df], axis=1)
data_iris.head()
```

```python
fig = px.scatter(data_iris.iloc[:,2], data_iris.iloc[:,3] ,color=iris.target)
fig.show()
```
