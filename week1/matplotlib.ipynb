```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
my_data = np.array([ ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], [12, 13, 9, 8, 7, 8, 8, 7, 6, 5, 8, 10], [14, 16, 11, 7, 6, 6, 7, 6, 5, 8, 9, 12]   ])
```

```python
np.shape(my_data)
```

```python
df = pd.DataFrame(my_data)
df
```

```python
traspose = df.transpose()  # exchanging the place
print(traspose)
```

```python
traspose.columns = ['Year', 'elec_01', 'elec_02']
traspose
```

```python
traspose.shape
```

```python
plt.plot(traspose.Year, traspose.elec_01, color='red', label='Elec 01', marker='o')

plt.plot(traspose.Year, traspose.elec_02, color = 'blue', label = 'Elec 02', marker = 'v')
plt.xlabel('Month')
plt.ylabel('Electricity Consumption')
plt.title('Building Consumption')
plt.legend(loc = 'upper center')

plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(traspose.Year, traspose.elec_01, color='red', label='Elec 01', marker='o')

plt.xlabel('Month')
plt.ylabel('Electricity Consumption')
plt.title('Building Consumption for customer 01')


plt.subplot(1,2,2)
plt.plot(traspose.Year, traspose.elec_02, color = 'blue', label = 'Elec 02', marker = 'v')
plt.xlabel('Month')
plt.title('Building Consumption for customer 02')
plt.show()


```

```python


plt.scatter(traspose.Year, traspose.elec_01, color = 'red', label = 'Customer 01')
plt.scatter(traspose.Year, traspose.elec_02, color = 'blue', label = 'Customer 02')
plt.xlabel('Month')
plt.ylabel('Electricity Consumption')
plt.title('Scatter Plot of Building Consumption')
plt.grid()
plt.legend(loc='best')
plt.show()


## scatter plotting
```

```python
#  plotting histogram

plt.hist(traspose.elec_01, bins=20, color = 'green')
plt.xlabel('Month')
plt.ylabel('Electricity Consumption')
plt.title('Histogram')
plt.show()
```

```python
## dobule bargraph

plt.bar(traspose.Year, traspose.elec_01, color = 'red', label = 'Customer 01')
plt.bar(traspose.Year, traspose.elec_02, color = 'blue', label = 'Customer 02')
plt.xlabel('Month')
plt.ylabel('Electricity Consumption')
plt.title('bar chart of Building Consumption')
plt.legend(loc='best')
plt.show()
```

```python
## bar graph with seperate analysis
bar_width = 0.4
months_b = np.arange(12)
plt.bar(months_b, traspose.elec_01,bar_width, color = 'blue', label =  'Customer 01')
plt.bar(months_b + bar_width, traspose.elec_02, bar_width, color = 'red', label = 'Customer 02')
plt.xticks(months_b + (bar_width)/12, ('Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'))
plt.xlabel('Month')
plt.ylabel('Electricity Consumption')
plt.title('bar chart of Building Consumption')
plt.legend(loc='best')
plt.show()
```
